```python
import csv
from collections import defaultdict

# Open the CSV file
with open('ODOP_List_Preprocessed.csv', 'r') as file:
    reader = csv.DictReader(file)

    # Initialize the hash map
    odop_words = defaultdict(int)

    # Extract and store each distinct word from the 'ODOP' column heading
    for row in reader:
        odop_heading = row['ODOP']
        for word in odop_heading.split():
            odop_words[word.casefold()] += 1

# Print the hash map
print(odop_words)
```
